```python
import pandas as pd
train=pd.read_csv("/kaggle/input/competitions/titanic/train.csv",index_col='PassengerId')
test=pd.read_csv("/kaggle/input/competitions/titanic/test.csv",index_col="PassengerId")

train_y=train.Survived
#test_y=test.Survived

#missing_col=[col for col in train.columns if train[col].isnull().any()]
#print(missing_col)

train_X=train.drop(['Name','Cabin','Ticket','Survived'],axis=1)
train_X.Embarked =train_X.Embarked.fillna("S")
train_X=pd.get_dummies(train_X,columns=['Sex','Embarked'])

#missing_col_test=[col for col in test.columns if test[col].isnull().any()]
#print(missing_col_test)
test_X=test.drop(["Name","Cabin","Ticket"],axis=1)
test_X=pd.get_dummies(test_X,columns=['Sex','Embarked'])


from sklearn.impute import SimpleImputer
impute=SimpleImputer(strategy='median')
train_X['Age']=impute.fit_transform(train_X[['Age']])

test_X['Age']=impute.transform(test_X[['Age']])

fare_imputer=SimpleImputer(strategy='median')
fare_imputer.fit(test_X[['Fare']])
test_X['Fare']=fare_imputer.transform(test_X[['Fare']])

from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=1)
model.fit(train_X,train_y)


predictions = model.predict(test_X)

output = pd.DataFrame({'PassengerId': test_X.index, 'Survived': predictions})
output.to_csv('submission.csv', index=False)

```
